```python
from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt
```

```python
lons = np.linspace(-179,180,360)
# careful, if lons are from 0 to 359 do something like lons_new = np.concatente([lons[180::-1],lons[0:180])
lons_classical = np.linspace(0,359,360,endpoint=True)
lons = np.concatenate([-360+lons_classical[181:],lons_classical[0:181]])
lats = np.linspace(-89,90,180, endpoint=True)
x,y = np.meshgrid(lons, lats)
grid = x,y
cmb = np.random.normal(size=(len(lats),len(lons))) # matrix
exposure = np.cos(x/10.) + np.sin(y/10.) # matrix from meshgrid
```

```python
# two ways, both do roughly same, contourf is more customizable, pcolormesh is faster
plt.figure()
m = Basemap(projection='hammer',lon_0=0) # or use 'moll'
m.drawparallels(np.arange(-90.,90.,30.),labels=[1,1,1,1])
m.drawmeridians(np.arange(0.,360.,30.))
x, y = m(*grid) # transform
c = m.contourf(x,y,exposure,30,cmap=plt.cm.jet)
m.colorbar(location='bottom', pad='10%')
plt.title('Exposure map (contours)')

plt.figure()
m = Basemap(projection='hammer',lon_0=0) # or use 'moll'
m.drawparallels(np.arange(-90.,90.,30.),labels=[1,1,1,1])
m.drawmeridians(np.arange(0.,360.,30.))
x, y = m(*grid) # transform
m.pcolormesh(x,y,cmb)
m.colorbar(location='bottom', pad='10%')
plt.title('Exposure map (pcolormesh)')
```
